```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/2023マリノス観客分析用.csv")
df
```

```python
df
```

```python
# prompt: NaNを０に置き換える

df.fillna(0, inplace=True)
df = df.drop(["日付（メモ）","試合前順位"],axis=1)
```

```python
# prompt: 入場者数、日付（メモ）以外の列名のリスト

cols = list(df.columns)
cols.remove('入場者数')
```

```python
x = df[cols]
y = df['入場者数']
```

```python
cols
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

model_lr = LinearRegression()
model_lr.fit(x, y)
```

```python
print(model_lr.coef_)
print(model_lr.intercept_)
print(model_lr.score(x, y))
```
